```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_simulation(initial_price, expected_return, volatility, days, simulations):
    dt = 1 / 252  # Convert to shares of the year (trading days per year)
    returns = np.exp((expected_return - 0.5 * volatility ** 2) * dt +
                     volatility * np.sqrt(dt) * np.random.normal(size=(days, simulations)))
    price_paths = initial_price * np.cumprod(returns, axis=0)

    return price_paths

# model parameters
initial_price = 40.0
expected_return = 0.16
volatility = 0.2
days = 252
simulations = 3


# Start simulation
price_paths = monte_carlo_simulation(initial_price, expected_return, volatility, days, simulations)

# Visualize results
plt.figure(figsize=(10, 6))

for i in range(simulations):
    plt.plot(price_paths[:, i], label=f'Simulation {i + 1}')

plt.title('Monte Carlo Simulation of Stock Prices')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()


```

The simulation is based on the assumption that percentage changes in stock prices over short periods are normally distributed. The key parameters used in the model include the expected return on the stock, its volatility, and the length of the time intervals.

Model Formulation:
The model is formulated using the lognormal property of stock prices, as outlined in Chapter 14 of the finance textbook. The expected return per year (µ) and volatility (σ) of the stock are essential inputs. The simulation generates future stock prices by simulating logarithmic returns and exponentiating the results. The standard deviation of the log returns is determined by the volatility parameter.

Transforming logarithmic prices into actual stock prices.
